```python
custlist_df['gender'].replace('U', pd.np.nan, inplace=True)
```

```python
#plotting gender against the mean value to get to know what gender has a higher value
grouped = custlist_df.groupby('gender')['Value'].mean()
print(grouped)
```

```python

sns.barplot(x=grouped.index, y=grouped.values)

plt.title('Mean Value by Gender')
plt.ylim(0, grouped.max()*5)
plt.show()
```
